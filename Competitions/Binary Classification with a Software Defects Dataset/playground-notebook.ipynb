## Dealing with outliers and missing/repeated values
First, let's check for any missing values

```python
msno.matrix(train, figsize=(15,5))
```

The MSNO library show's that there ain't no missing values on the dataset.

Now, let's remove duplicated rows from the dataset

```python
train = train.drop_duplicates(inplace=False)
```

After that, we're going to check and cut some outliers from some features, increasing our model performance and our visualization from now on

```python
var = train.columns.values

i = 0

sns.set_style('whitegrid')
plt.figure()
fig, ax = plt.subplots(8,3,figsize=(16,28))

for feature in var:
    i += 1
    plt.subplot(8,3,i)
    try:
        sns.scatterplot(x='id', y=feature, data=train)
    except TypeError:
        continue
    plt.xlabel(feature, fontsize=12)
    locs, labels = plt.xticks()
    plt.tick_params(axis='both', which='major', labelsize=12)
clear_output()
plt.show();
```

## Features Evaluation
Now, we're going to check the correlation between features/'defects' and features/features, so we can proceed with some **feature engineering** methods and **prioritize** some of them

```python
plt.pie([len(train[train['defects']==True]), len(train[train['defects']==False])], labels=['True', 'False'], autopct = '%.2f%%')
clear_output()
```

```python
plt.figure(figsize = (12,12))
plt.title('CorrelaÃ§Ã£o entre features (Pearson)')
corr = train.corr()
sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns,linewidths=.1,cmap="Reds")
plt.show()
```

```python
var = train.columns.values

i = 0
t0 = train.loc[train['defects'] == False]
t1 = train.loc[train['defects'] == True]

sns.set_style('whitegrid')
plt.figure()
fig, ax = plt.subplots(8,3,figsize=(16,28))

for feature in var:
    i += 1
    plt.subplot(8,3,i)
    sns.kdeplot(t0[feature], bw=0.5,label="Class = False")
    sns.kdeplot(t1[feature], bw=0.5,label="Class = True")
    plt.xlabel(feature, fontsize=12)
    locs, labels = plt.xticks()
    plt.tick_params(axis='both', which='major', labelsize=12)
clear_output()
plt.show();
```

#### Those 2 plots above show a decent selectivity in terms of distribution for the variable **l**, but still a high correlation between **defects** and the rest of the features

For that, I'll try the following approaches:

- **Create Interaction Terms:** If two features are highly correlated, creating an interaction term (multiplying the two features) might help the model capture their combined effect.
- **Binning or Discretization:** Convert continuous features into categorical features by binning them. This can sometimes capture non-linear relationships more effectively.

First, let's see the distribution in those continuos variables after binning them

## 1.0 - Binning variables

```python
data_backup = train
bins = 3
data_backup['timegroup'] = pd.qcut(data_backup['loc'], bins)

fig, ax = plt.subplots(figsize=(20,7))
ax11 = fig.add_subplot(221)  # add subplot into first position in a 2x2 grid (upper left)
ax12 = fig.add_subplot(223, sharex=ax11)  # add to third position in 2x2 grid (lower left) and sharex with ax11
ax13 = fig.add_subplot(122)
#sns.scatterplot(ax = ax11, x='id', y='loc', hue='defects', data=data_backup, edgecolor=None, s=3).set(title='Original Variavel "loc"')
#sns.scatterplot(ax = ax12, x='id', y='loc', hue='timegroup', data=data_backup, edgecolor=None, s=3).set(title='Binned Variavel "loc"')
t0 = train.loc[train['defects'] == False]
t1 = train.loc[train['defects'] == True]

sns.kdeplot(t0['loc'], bw=0.5, label = 'Class = False', ax=ax11)
sns.kdeplot(t1['loc'], bw=0.5,label="Class = True", ax = ax11)

sns.barplot(ax = ax13, x='timegroup', y='loc', data=data_backup).set(title=f'Binned Variavel "loc" - {bins} bins')

## Falta separar a variÃ¡vel loc em bins e checar nova correlaÃ§Ã£o
clear_output()
plt.show()
```
