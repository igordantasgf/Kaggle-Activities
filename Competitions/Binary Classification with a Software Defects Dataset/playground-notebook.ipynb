## Dealing with outliers and missing/repeated values
First, let's check for any missing values

```python
msno.matrix(train, figsize=(15,5))
```

The MSNO library show's that there ain't no missing values on the dataset.

Now, let's remove duplicated rows from the dataset

```python
print(f"Before : {len(train)}")
train = train.drop_duplicates(inplace=False)
print(f"After : {len(train)}")
```

After that, we're going to check and cut some outliers from some features, increasing our model performance and our visualization from now on

```python
print(f'Rows before cutting data: {len(train)}')
```

```python
train = train[train['loc'] < 600]
train = train[train['v(g)'] < 100]
train = train[train['ev(g)'] < 70]
train = train[train['iv(g)'] < 100]
train = train[train['n'] < 3000]
train = train[train['v'] < 18000]
train = train[train['l'] < 0.8]
train = train[train['i'] < 400]
train = train[train['d'] < 200]
train = train[train['e'] < 4000000]
train = train[train['b'] < 8]
train = train[train['t'] < 100000]
train = train[train['lOBlank'] < 100]
train = train[train['locCodeAndComment'] < 18]
train = train[train['uniq_Op'] < 70]
train = train[train['uniq_Opnd'] < 200]
train = train[train['total_Op'] < 1300]
train = train[train['total_Opnd'] < 1000]



print(f'Rows after cutting data: {len(train)}')
```

```python
var = train.columns.values

i = 0

sns.set_style('whitegrid')
plt.figure()
fig, ax = plt.subplots(8,3,figsize=(16,28))

for feature in var:
    i += 1
    plt.subplot(8,3,i)
    try:
        sns.scatterplot(x='id', y=feature, data=train)
    except TypeError:
        continue
    plt.xlabel(feature, fontsize=12)
    locs, labels = plt.xticks()
    plt.tick_params(axis='both', which='major', labelsize=12)
clear_output()
plt.show();
```

## Features Evaluation
Now, we're going to check the correlation between features/'defects' and features/features, so we can proceed with some **feature engineering** methods and **prioritize** some of them

```python
plt.figure(figsize = (12,12))
plt.title('Correlação entre features (Pearson)')
corr = train.corr()
sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns,linewidths=.1,cmap="Reds")
plt.show()
```

```python
var = train.columns.values

i = 0
t0 = train.loc[train['defects'] == False]
t1 = train.loc[train['defects'] == True]

sns.set_style('whitegrid')
plt.figure()
fig, ax = plt.subplots(8,3,figsize=(16,28))

for feature in var:
    i += 1
    plt.subplot(8,3,i)
    sns.kdeplot(t0[feature], bw=0.5,label="Class = False")
    sns.kdeplot(t1[feature], bw=0.5,label="Class = True")
    plt.xlabel(feature, fontsize=12)
    locs, labels = plt.xticks()
    plt.tick_params(axis='both', which='major', labelsize=12)
clear_output()
plt.show();
```

#### Those 2 plots above show a decent selectivity in terms of distribution for the variable **l**, but still a high correlation between **defects** and the rest of the features

For that, I'll try the following approaches:

- **Create Interaction Terms:** If two features are highly correlated, creating an interaction term (multiplying the two features) might help the model capture their combined effect.
- **Binning or Discretization:** Convert continuous features into categorical features by binning them. This can sometimes capture non-linear relationships more effectively.

First, let's see the distribution in those continuos variables after binning them

## 1.0 - Binning variables

```python
data_backup = train

def binning_plot(variavel, bins):

    data_backup['timegroup'] = pd.qcut(data_backup[variavel], bins, labels=False, retbins=False)

    fig, ax = plt.subplots(1,3,figsize=(18,4))

    t0 = train.loc[train['defects'] == False]
    t1 = train.loc[train['defects'] == True]
    
    sns.kdeplot(x=variavel, bw=0.5, data=t0, ax=ax[0]).set(title='Distribution befode Binning')
    sns.kdeplot(x=variavel, bw=0.5, data=t1, ax=ax[0]).set(title='Distribution befode Binning')

    #data_backup[data_backup['defects']==0]['loc'].plot(kind='bar', alpha=0.5, ax=ax12)
    sns.histplot(ax = ax[1], x='timegroup', hue="defects", multiple="stack", data=data_backup, label='Distribution befode Binning').set(title=f'Distribution with {bins} bins')

    sns.boxplot(x='defects', y='timegroup', data=data_backup, ax=ax[2]).set(title=f'Distribution with {bins} bins')  
    #sns.boxplot(y='timegroup',  data=t0, ax=ax[2]).set(title=f'Distribution with {bins} bins')
    #sns.boxplot(y='timegroup',  data=t1, ax=ax[2]).set(title=f'Distribution with {bins} bins')

    clear_output()
    plt.show()
    
binning_plot('loc',20)
```
